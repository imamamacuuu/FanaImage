#**BRUTE FORCE**

- Brute: Kasar
- Force: Kekuatan
- Brute Force: Kekuatan yang kasar ( tidak melibatkan kecerdasan )

Brute Force: Metode penyelesaian dengan mencoba semua kemungkinan satu per satu hingga menemukan solusi yang benar.

##Ciri-ciri brute force
- lebih mengandalkan kemampuan prosesor untuk menyelesaikan masalah daripada mencari solusi yang lebih efisien
- sederhana
- Langsung dan Just do it!

##Contoh permasalahan yang diselesaikan menggunakan brute force
1. Menghitung an (a > 0, n a nonnegative integer)
2. menghitung n!
3. Perkalian matriks
4. searching
5. Sorting
6. Password Cracking
7. dll


```python
def faktorial(n):
    hasil = 1
    for i in range(1, n + 1):
        hasil *= i
        print(hasil)
    return hasil

# Contoh penggunaan
angka = 5
print(f"Faktorial dari {angka} adalah {faktorial(angka)}")
```

# **Searching Menggunakan BruteForce**

```python
#MENCARI INDEX DARI ANGKA 21

list = [3, 5, 12, 15, 21, 22, 34, 36]

key = 21
idx = 0
for number in list:
  print(f"index: {idx}, number = {number}")
  if number == key :
    print(f"angka {key} terdapat pada index ke-{idx} pada list")
    break
  else:
    idx += 1
```

```python
#CARA LAIN (FUNCTION)
percobaan = 0

list = [3, 5, 12, 15, 21, 22, 34, 36]
key = 21

def cari(list, key):
  idx = 0
  for number in list:
    global percobaan
    percobaan += 1
    print(f"Percobaan ke [ {percobaan} ]")
    if number == key :
      return idx
      break
    else:
      idx += 1

cari(list, key)
```

```python
#CARA LAIN (MEMANFAATKAN REKURSIF)
percobaan = 0

def findIdx(list_soal, search, idx=0):
    global percobaan
    percobaan += 1
    print(f"Percobaan ke [ {percobaan} ] index [ {idx} ] value [ {list_soal[idx]} ]")

    if idx >= len(list_soal):  # Jika index melebihi panjang list, elemen tidak ditemukan
        return -1

    if list_soal[idx] == search:
        return idx
    else:
        return findIdx(list_soal, search, idx + 1)

# Contoh Penggunaan
result = findIdx([3, 5, 12, 15, 21, 22, 34, 36], 21)
print(f"Ditemukan di index ke [ {result} ]")
```

---

## **Apakah Brute Force Efektif?**
Tentu **tidak selalu efektif**. Misalnya, jika kita memiliki sebuah **List dengan panjang 1000**, Brute Force akan memakan waktu yang sangat lama karena kompleksitas waktunya yang tinggi (O(N). Hal ini membuat Brute Force tidak cocok untuk masalah dengan skala besar.

---

## **Solusi: Menggunakan Algoritma yang Lebih Efisien**
Untuk mengatasi kelemahan Brute Force, kita dapat menggunakan algoritma yang lebih efisien, seperti:

1. **Jump Search**:  
   Meningkatkan efisiensi pencarian dengan melompati beberapa elemen sekaligus.

2. **Binary Search**:  
   Membagi ruang pencarian menjadi dua bagian secara berulang, sehingga mengurangi kompleksitas waktu menjadi O(log n).

3. **Hashing**:  
   Menggunakan struktur data seperti dictionary atau hash table untuk mengurangi waktu pencarian menjadi O(1) dalam kasus terbaik.

```python
# Contoh sederhana Searching Menggunakan Jump Search (Versi Sederhana)

# MENCARI INDEX DARI KEY DI LIST
# list = [3, 5, 12, 15, 21, 22, 34, 36]
# list = [5, 7, 12, 16, 54, 66, 72, 93, 108, 119, 144, 157, 163, 171, 177, 184, 186, 195, 203, 213, 217, 222, 229, 235, 255, 271, 279, 283, 289, 291, 293, 305, 311, 355, 363, 364, 366, 369, 378, 391, 415, 421, 423, 425, 427, 437, 447, 453, 455, 487, 489, 494, 517, 540, 547, 551, 553, 568, 574, 580, 581, 594, 601, 606, 615, 617, 632, 641, 653, 660, 664, 666, 691, 743, 753, 767, 768, 769, 794, 795, 799, 806, 807, 814, 817, 838, 844, 855, 868, 869, 875, 901, 907, 909, 914, 929, 955, 957, 986, 998]

key = 211
idx = 0
step = 4  # Langkah lompatan

percobaan = 0

# Loop utama untuk Jump Search
print(f"\nJump Search\n")

while idx < len(list):
    next_idx = min(idx + step, len(list) - 1)  # Batasi agar tidak melebihi panjang list
    percobaan += 1
    print(f"Percobaan ke [ {percobaan} ] index [ {next_idx} ] value [ {list[next_idx]} ]")
    if list[next_idx] >= key:
        # Lakukan linear search di blok saat ini
        for i in range(idx, next_idx + 1):
            if list[i] == key:
                percobaan += 1
                print(f"Percobaan ke [ {percobaan} ] index [ {i} ] value [ {list[i]} ]")
                print(f"Angka {key} ditemukan di indeks: {i}")
                break
        else:
            print(f"Angka {key} tidak ditemukan.")
        break
    idx += step  # Lompat ke blok berikutnya
else:
    print(f"Angka {key} tidak ditemukan.")


# Perbandingan dengan Linear Search
print(f"\nLinear Search (BruteForce)\n")

percobaan = 0
def cari(list, key):
  idx = 0
  for number in list:
    global percobaan
    percobaan += 1
    print(f"Percobaan ke [ {percobaan} ] index [ {idx} ] value [ {list[idx]} ]")
    if number == key :
      print(f"Angka {key} ditemukan di index: {idx}")
      break
    else:
      idx += 1
  else:
    print(f"Angka {key} tidak ditemukan.")

cari(list, key)
```
